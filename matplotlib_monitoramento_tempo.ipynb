```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
temperatura_por_dia_da_semana 

#calculando a média de temperatura por dia da semana
```

```python
temperatura_por_dia_da_semana = df.groupby('dia_da_semana')['temperatura'].mean()
nome_dias = ['Domingo', 'Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta', 'Sábado']
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana[nome_dias]
temperatura_por_dia_da_semana

#ordenando os dias da semana
```

```python
fig = plt.figure(figsize=(10,4))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))
cores = ['black', 'red', 'green', 'blue', 'yellow', 'orange', 'magenta']

eixo.bar(indice, temperatura_por_dia_da_semana, color=cores, edgecolor='k')
eixo.set_title('Temperatura por dia da semana', fontsize=16)
eixo.set_ylabel('Temperatura', fontsize=16)
eixo.set_xlabel('Dias da semana')
eixo.set_xticks(indice)
eixo.set_xticklabels(nome_dias)

##Criando um gráfico de barras com as médias de temperatura por semana
```

```python
fig = plt.figure(figsize=(8,5))
eixo = fig.add_axes([0,0,1,1])

indice = range(len(temperatura_por_dia_da_semana))
cores = ['black', 'red', 'green', 'blue', 'yellow', 'orange', 'magenta']
explodir = [0.2, 0 , 0 , 0, 0, 0, 0.2]

eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index, autopct='%.1f%%', explode=explodir, shadow=True)
eixo.set_title('Temperatura por dia da semana', fontsize=16)

##Criando um gráfico de pizza com as médias de temperatura por semana
```
